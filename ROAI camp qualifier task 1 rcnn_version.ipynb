```python
from PIL import Image
class DF(Dataset):
    def __init__(self, df, path, trans=None,is_test=False):
        self.df = df
        self.path = path
        self.trans = trans
        self.is_test = is_test
    def __len__(self):
        return len(self.df)
    def __getitem__(self, idx):
        img_path = os.path.join(self.path, self.df["filename"][idx])
        img = Image.open(img_path).convert("RGB")
        w, h = img.size
        x_scale = 224 / w
        y_scale = 224 / h
        img = self.trans(img)
        if self.is_test:
            return img
        boxes = ast.literal_eval(self.df["gt_boxes"][idx])
        boxes = torch.tensor(boxes, dtype=torch.float32)
        boxes[:, [0, 2]] *= x_scale
        boxes[:, [1, 3]] *= y_scale
        labels = torch.ones((boxes.shape[0],), dtype=torch.int64)
        target = {
            "boxes": boxes,
            "labels": labels,
            "image_id": torch.tensor([idx])
        }
        return img, target

from torchvision import transforms as T
train_trans = T.Compose([
    T.Resize((224, 224)),
    T.ColorJitter(0.2, 0.2, 0.2, 0.4),
    T.GaussianBlur((3, 3)),
    T.ToTensor(),
    T.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])
test_trans = T.Compose([
    T.Resize((224, 224)),
    T.ToTensor(),
    T.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])
def collate_fn(batch):
    return tuple(zip(*batch))
train_df= DF(train_ds, train_imgs, trans = train_trans, is_test= False)
test_df= DF(train_ds, train_imgs, trans = test_trans, is_test= True)
train_dl = DataLoader(train_df,batch_size=16,shuffle=True,num_workers=4,pin_memory=True,collate_fn=collate_fn)
test_dl = DataLoader(test_df,batch_size=16,shuffle=False,num_workers=4,pin_memory=True,collate_fn=collate_fn)
device = ("cuda" if torch.cuda.is_available() else "cpu")
```

```python
from torchvision.models.detection.faster_rcnn import FastRCNNPredictor
class MODEL(nn.Module):
    def __init__(self):
        super().__init__()
        self.model = fasterrcnn_resnet50_fpn(weights="DEFAULT")
        inf = self.model.roi_heads.box_predictor.cls_score.in_features
        self.model.roi_heads.box_predictor = FastRCNNPredictor(inf,2)
    def forward(self, x, y=None):
        return self.model(x, y)

model = MODEL().to(device)
```

```python
opt = torch.optim.AdamW(model.parameters(), lr = 1e-4, weight_decay= 2e-4)
for e in range(10):
    t_loss = 0
    model.train()
    for xs, ys in train_dl:
        xs = [x.to(device) for x in xs]
        ys = [{k: v.to(device) for k, v in y.items()} for y in ys]
        ld = model(xs, ys)
        loss = sum(loss for loss in ld.values())
        t_loss += loss.item()
        opt.zero_grad()
        loss.backward()
        opt.step()
```

```python
test_preds = []
model.eval()
with torch.no_grad():
    for batch in test_dl:
        xs = list(x.to(device) for x in batch)
        out = model(xs)   
        test_preds.extend(out)
test_preds= np.concatenate(test_preds, axis  = 1)
len(test_preds), len(test_df)
```
